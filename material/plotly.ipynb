```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
years=[i for i in range(2018,2023)]
sales=[100,200,178,190,130]
```

```python
fig=go.Figure(data=go.Scatter(x=years,y=sales))
fig.show()
```

```python
fig=go.Figure(data=go.Scatter(x=years,y=sales,name="hello"))
fig.update_layout(
    title="Sales v/s Years",
    xaxis_title="Years",
    yaxis_title="sales",
   
    showlegend=True
)
fig.show()
```
