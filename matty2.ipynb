```python
# i am learining matplotlib
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()   #this makes grid in the graph 
plt.show()   #this shows the graph
# This code generates a simple sine wave plot using matplotlib.
```

```python
x=[1,2,3,4,5]
y=[2,3,4,5,6]
plt.plot(x, y)
plt.title("Line Plot")  #this is title of the graph
plt.xlabel("X-axis")    #this is x-axis label
plt.ylabel("Y-axis")    #this is y-axis label
plt.grid()   #this makes grid in the graph  
plt.show()   #this shows the graph
```

```python
df=pd.read_csv("homeprises.csv")
```

```python
df
```

```python
plt.scatter(df['area'], df['price'], color='green', marker='x')
plt.title("Home Prices")
plt.xlabel("Area (sq ft)")
plt.ylabel("Price ($)")
plt.grid()
plt.show()
```

```python
#3d plot
ax=plt.axes(projection='3d')
ax.scatter(df['area'], df['price'], df['bedrooms'], color='purple', marker='*')
plt.xlabel("Area (sq ft)")
plt.ylabel("Price ($)")
ax.set_zlabel("Bedrooms")
plt.show()
```
